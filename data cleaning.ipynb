```python
import pandas as pd
import numpy as np
pd.set_option("display.max_columns", None)
```

```python
credit_data = pd.read_csv('datasets/credit_train.csv')
credit_data.head()
```

```python
credit_data.info()
```

```python
credit_data.describe(include="all")
```

### Removing columns that are not important for our analysis

```python
credit_data = credit_data.drop(columns=['Loan ID', 'Customer ID', 'Months since last delinquent'])
credit_data.columns
```

We removed the columns 'Loan ID' and 'Customer ID' as they do not provide any insights; they are just random unique values. We dropped 'Months since last delinquent' since it contains a large proportion of NA values and it is not very relevant to our topic of analysis.

### Removing rows that have missing data

```python
credit_data = credit_data.dropna()
credit_data.head()
```

```python
credit_data.shape
```

### Checking for and removing duplicates

```python
credit_data.duplicated()
```

```python
credit_data = credit_data.drop_duplicates()
credit_data.shape
```

### Removing outliers/anomalous data

```python
credit_data = credit_data[credit_data['Current Loan Amount']!=99999999]
credit_data.shape
```

We noticed there is a placeholder of 99999999 for unknown 'Current Loan Amount', so we removed it from the dataset.

```python
credit_data = credit_data[credit_data['Credit Score']<=850]
credit_data = credit_data[credit_data['Credit Score']>=300]
credit_data.shape
```

We noticed there were outlier values for 'Credit Score'. The base FICOÂ® Scores range from 300 to 850, so we removed anything out of this range.

```python
Q3 = credit_data['Maximum Open Credit'].quantile(.75)
Q1 = credit_data['Maximum Open Credit'].quantile(.25)
credit_data = credit_data[credit_data['Maximum Open Credit']<=(Q3+1.5*(Q3-Q1))]
credit_data['Maximum Open Credit'].skew()
```

We noticed there were high outliers for 'Maximum Open Credit', with the maximum value (1.539738e+09) being much higher than the 3rd quartile (7.829580e+05). We used the Interquartile Range method (1.5 * IQR) to remove the higher bound outliers. Skewness of 0.8766170181905047 is moderately skewed and acceptable.

### Standardising names

```python
credit_data['Purpose'].replace('other', 'Other', inplace=True)
credit_data['Purpose'].replace('small_business', 'Small Business', inplace=True)
credit_data['Purpose'].replace('major_purchase', 'Major Purchase', inplace=True)
credit_data['Purpose'].replace('renewable_energy', 'Renewable Energy', inplace=True)
credit_data['Purpose'].replace('moving', 'Moving', inplace=True)
credit_data['Purpose'].replace('vacation', 'Vacation', inplace=True)
credit_data['Purpose'].replace('wedding', 'Wedding', inplace=True)
credit_data['Home Ownership'].replace('HaveMortgage', 'Have Mortgage', inplace=True)
```

We noticed in 'Purpose' the category names 'other' and 'Other' were not standardised, so we standardised it to 'Other'. We also standardised the other category names in 'Purpose' and 'Home Ownership for better organisation.

```python
credit_data.to_csv('datasets/cleaned_credit.csv', index=False)
```
